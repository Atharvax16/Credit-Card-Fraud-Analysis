```python
from sklearn.model_selection import KFold

def target_encode_kfold(
    train_df,
    test_df,
    col,
    target = 'is_fraud',
    n_splits = 5,
    smoothing = 10
):
    
    global_mean = train_df[target].mean()
    kf = KFold(n_splits = n_splits,shuffle = True,random_state = 42)
    
    train_encoded = pd.Series(index = train_df.index,dtype = float)
    
    for train_idx,val_idx in kf.split(train_df):
        X_tr = train_df.iloc[train_idx]
        X_val = train_df.iloc[val_idx]
        
        stats = X_tr.groupby(col)[target].agg(['mean','count'])
        
        smooth = (
            (stats['count'] * stats['mean'] + smoothing * global_mean) /
            (stats['count'] + smoothing)
        )
        
    train_encoded.fillna(global_mean, inplace=True)
    stats_full = train_df.groupby(col)[target].agg(['mean', 'count'])
    smooth_full = (
        (stats_full['count'] * stats_full['mean'] + smoothing * global_mean) /
        (stats_full['count'] + smoothing)
    )

    test_encoded = test_df[col].map(smooth_full)
    test_encoded.fillna(global_mean, inplace=True)

    return train_encoded, test_encoded
```
